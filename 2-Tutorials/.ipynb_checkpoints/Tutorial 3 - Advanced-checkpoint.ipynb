# 1. Backgroud on Machine Learning: _What is Machine Learning?_

In the world of analytics and specifically Data Science, _"Machine Learning"_ is so ubiquitos and a big buzzword thrown all over. Sometimes in the context of _"We use ML in our (insert product)!"_ or _"Just use ML and you'll get the answer",_ being almost this esoteric concept in Data Science often associated as part of AI.

A better way of viewing Machine Learning is as the union of the concepts of computer programming (Beginner Notebook) and statistical concepts (Intermediate Notebook) in Data Science. The primary idea to remember is that we are building models out of data, where our models "learn" or become tuned from data and then can make predictions on similar but never before seen data.  


## 1.1 Types of Machine Learning Models

There are a large variety of Machine Learning models available for us work with, but it is important to always remember that our problem or question at hand will dictate the type of models we can use. In other words, _don’t try to paint a wall with a hammer_ or _screw on a shelf with a saw_ , each tool has a best use scenario. It might sound enticing to use a fancy sounding model say “K-Means” or “Support Vector Machine” for your research, but if you are trying to predict effects of drug dosage on how fast a cancer metastasize, you might be using the wrong tool for the job.

Let’s first discuss the two main types Machine Learning algorithms are categorized into: [_Supervised_](https://en.wikipedia.org/wiki/Supervised_learning) and [_Unsupervised_](https://en.wikipedia.org/wiki/Unsupervised_learning) .

### 1.1.1 Supervised Machine Learning

Recall that Machine Learning is our models “learning” from our data, hence _Supervised Machine Learning_ would be a model that “learns” from data that has some labels already attached to it. Some of the models that fall under this type of ML are _Regression_ and _Classification_ , such as Linear Regression model with continuous data to predict a continuous value (age, dosage, time, etc) or Logistic Regression for classifying values (sex, smoker, education level, etc).  If our data has labels already attach to them and we want to figure out the relation of a target variable to these labels, then we are dealing with a Supervised Learning problem.

### 1.1.2 Unsupervised Machine Learning

Given that Supervised Learning requires labels for our data, _Unsupervised Machine Learning_ is the name given to models that use unlabeled data, which mainly means we “omit” labels that might exist in our data, allowing the model to learn patterns from some features. These types of models tend to be used for _Clustering_ or for _Dimension Reduction_ purposes.

## 1.2 Key Basic ML Ideas and Concepts

Machine Learning like any other specialized field has its own language and key concepts that are ubiquitous across all sections of the discipline. We will touch upon a few of these as we work through the tutorial and the examples, as they will be best address there. For now, we want to point out 4 of the most common used terminalogy and concepts so we have a better understanding as we start the tutorial.

### 1.2.1 Training and Testing Data Split

One of the key concept within machine learning is that we are _"training our statistical model on data, to then make predictions"_ , key here is the idea that we need to have data to ___train___ our models. Machine Learning algorithms are mathematical algorithms and are closely related to statistics, and as such they take in the data and try to find some optimize way to solve the algorithm, hence it is being _trained_ by data. We feed the data, our models are trained, and then we check how well our model does at predicting what we want. 

_Ok data is 'training' our models. But why do we need to have 'multiple separate' datasets?_

For example:
> We have a dataset of house listings, with features of house sizes, neighborhood, year built, number of rooms and bathrooms, and price sold. We want to predict how much a house will sell for given the other features. So we would train our model on our data, feeding all the features and the target to predict ('sale price') for our model. We would then want to find out how our model performs, how well it matches real life sale prices.  
__How can we check?__  
_We could take some of the listings in our data and use them and try to predict their sale price with our trained model, great, what's the problem?_
 

### 1.2.2 Overfitting, Underfitting, Bias-Variance Tradeoff

When we assess how well a _trained_ model is performing at a task by using the same data used to train the model, we run into the problem of ___Overfitting___ .

> In statistics, __Overfitting__ is the production of an analysis which corresponds too closely or exactly to a particular set of data, and may therefore fail to fit additional data or predict future observations reliably.  
-OxfordDictionary.com

So, _Why do we need to have __independent Training and Testing__ datasets?_ We'll if we don't, we run into the problem of not knowing how we are overfitting and properly evaluate how our model is performing and if it is not “learning” the overall features of the data, rather than focusing on  small fluctuations or "noise" of the data.

Similarly, if we can _Overfit_ a model, a model can also be ___Underfitted___, meaning that instead of being caught in small fluctuations of data, now it misses the overall trend of the data. Underfitting occurs when our model has been trained on too small of a dataset for our model to capture enough information.

Both of these are part of what is known as ___The Bias-Variance Tradeoff___ . We try to accurately train models, without being caught in the noise _(Overfitting - High variance and low bias)_ and that don’t miss out on the trends _(Underfitting - High bias and low variance)_ , in the training data to be used in our test data.

We can address these in a few ways:

Overfitting | Underfitting 
--------------|-------------
Increase Training Data | Reduce/Remove Noise  
Change the type of Loss function | Increase model complexity 
Reduce Features (Selection/engineering) | Increase Features (Selection/engineering)

these are just some suggestions we can go about in trying to address the tradeoff, but as we can tell there is a sweetspot that we are trying to compromise on so that our model can have a _good_ fit once we evaluate our model on new data.

### 1.2.3 Feature Engineering 

Feature engineering is one of the most important ways in which we have control over the complexity and accuracy of our models, that is to say also addressing the _Bias-Variance Tradaoff_ . So as we know, our datasets can usually be broken into _features_ to model and a _target_ variable to be predicted by a model. So immediately we can infer how important features are for how well a model will perform. 

It is important to address what it is that we are trying to accomplish with our model, as this can help us get started with “engineering” the best or most useful features. We need to be able to identify the usefulness and importance a feature can have in addressing our goal, we probably would not think to add a patient’s “favorite energy drink flavor” as an appropriate feature for predicting blood pressure. So appropriate _feature selection_ is a key aspect in the data processing step developing a good model.

_Feature engineering_ is the process of crafting domain specific features from the ones “naturally” available in the data. These might be us carefully constructing features from open-ended survey answer or creating some ratios that better encompass aspects of the data than the original features. For example, we can create _binary features (0 or 1/ yes or no)_ features if data matches a condition (i.e., is patient diabetic, gender, age group, etc).  Similarly, we can create features like percent of time systolic BP is high during workout or pregnancy. Here is where we want to acknowledge that feature engineering is part of the _art in the craft_ of machine learning.

For clarity, we want to maximize our _Objective Function_ and this is accomplished by minimizing our _Loss Function_ . 

We have quite a few options for _Loss Functions_ to choose from depending on what we are trying to accomplish with out models, but we can for simplicity also break them into _Classification and Regression Loss Functions_ . A few examples of these are:

Classification | Regression 
--------------|-------------
Log Loss | Mean Square Error  
Hinge Loss | Mean Absolute Error 
Exponential Loss | Huber Loss

Some of the most used and even _default_ functions in prepackage machine learning models (like the ones we'll use in the tutorial) are _Log Loss_ and _Mean Square Error_ .

As we discussed earlier, evaluating how well our models perform on _”New or Unseen”_ data, is a crucial step in making sure our model is performing as we expect. There are a variety of ways in which we can accomplish this as we can imagine, but again we can split them into _Classification and Regression Evaluation Metrics_ .
For Classification we have

### 1.3.1 Classification Models

__Classification__ in Machine Learning refers to models that predict a label(s) for the given data, and they pertain to _Supervised Learning_ since a label is needed. 

In _classification_ problems, we usually have a descriptive _'Class Label'_ we are attempting to predict from the features in our data. This requires the data used to trained our model to have sufficiently many samples with features and labels for our model to learn from. These classification labels can be binary, Multi-Class, or Imbalanced. 

These models calculate the best mapping of the features to the predicted labels, which can be then evaluated based on how well the _predicted_ labels performed compared to the _actual_ labels from a test or holdout set.

### Logistic Regression

Probably the simplest and most used classification model, _Logistic Regression_ is most often used as a starting 'baseline' model to then compared the performance of more complex models.

Logistic Regression works by implementing what is known as [_Sigmoid Function_](https://en.wikipedia.org/wiki/Sigmoid_function):
$$
S(x) = \frac{1}{1+e^{-x}}
$$
> $S(x)$: is an output between 0 and 1 for probability estimation  
$x$: input for our function  

Given the set of features in our data, once fed into the function we obtained a probability estimate which can then be transformed into a 0 or 1 depending on a _Decision Boundary_ . The boundary is a 'cutoff' value where we dictate that a specific sample is either 0, if they are under the boundary value, or 1, if over the boundary value.


#### Logistic Regression Pros and Cons
Pros | Cons
--------- | -----------
Easy to implement and use | Performance highly dependent on feature selection and feature engineering
Simple, efficient, and does not require too many computational resources | Not very powerful, can be easily outperformed by more complex models
Highly interpretable | can't solve non-linear problems, best use if categories can be "separated by a line"
Little or No tunning needed | Can only predict categorical outcome and vulenrable to overfitting



### Decision Tress and Random Forests

_Random Forests_ are basically an emsemble of _Decision Trees_ , so they are some of the most used and also powerful models in the area of supervised machine learning as they provide better accuracy, stability, and ease of interpretation.

The Decision Tree model works around the concepts of [_Information Gain_ and _Entropy_](https://en.wikipedia.org/wiki/Information_gain_in_decision_trees):

> $$
Entropy: H(X) = - \sum_{i} p_i(x) (log_2 * p_i(x))
$$
$$
Information \ Gain: IG(T, a) = H(T) - \sum_{v\in values(a)} P_a(v) * H(T_a(v)) 
$$
or can also be described as  
$$
IG(T, a) = H(T) - H(T|a)
$$
_Information Gain_ is the entropy of $T$ minus the entropy of the values of $T$ given $a$.

In the end the goal of the tree models is to maximize the _Information Gain_ while minimizing the _entropy_ .

#### Tree models Pros and Cons
Pros | Cons
--------- | -----------
Easy to use and good predictive power | Can Overfit
It is lenient on data preprocessing | shows bias in favor for features with more levels
Allows for various parameters to tune for better performance | many tunning paramenters can also become complex
Provides feature importance in the model | Can feel like a 'blackbox' at times

### Support Vector Machines(*SVM*)

_Support Vector Machines_ are supervised learning models that work by computing an optimal hyperplane that allows for a clear separation of the groups in the data.

SVM primarily focuses on trying to maximize the margin of the hyperplane that allows for the separation, and this is accomplish by focusing on maximizing the [_Hinge Loss_](https://en.wikipedia.org/wiki/Hinge_loss) function:
> $$
c(x, y, f(x)) = \left\{
\begin{array}{ll}
      0 & if \ y*f(x) > 1 \\
      1 - y*f(x) & else \\
\end{array} 
\right.
$$


#### SVM Pros and Cons
Pros | Cons
--------- | -----------
Good for data with clear separation | Does not perform well if there is much noise, poor separation
Effective in high dimensional spaces | Not that efficient with large data sets
Effective when there are more features than samples | No probabilistic explanation for the classification

### 1.3.2 Clustering Models

__Clustering__ is the category of ML tools that deals with grouping or partitioning objects, and these pertain to _Unsupervised Learning_ since we allow the models to make sense of the data.

Datasets typically do not contain a measure that you should group or partition by directly. Similar to the 'loss' functions used in other ML models, clustering algorithms use various derived metrics which are supposed to capture the difference, or 'dissimilarity', between observations. The goal of clustering is to identify subsets of data for which the dissimilarity metric is low within a subset and higher between the subsets.

Once a satisfactory clustering is achieved by a given model, it can be used to assign new observations (e.g. new samples of gene expressions) to the established groupings. A common example of this is image classification. Once you have a working model that groups photos into 'cat' and 'dog' clusters, you can label new photos as being 'cat' or 'dog' based on which cluster they would be assigned to.

The next few sections offer a brief explanation of a few different clustering algorithms, and the strengths and weaknesses of each. **Avoid just using a random clustering algorithm in any situation.**

### K-Means

Perhaps the most straightforward clustering algorithm is *k-means*.

The metric this method focuses on is euclidian distance between points. For reference:

> Euclidian Distance
>
> $d(p, q) = \sqrt{\sum_{i=0}^{n}{(q_{i} - p_{i})^{2}}}$
>
> where **p** and **q** are points in an **n** dimensional euclidian space
>
> $q_{i}, p_{i}$ are the components of the corresponding vectors for each point

The algorithm is as follows:

>1. Pick K random points as cluster 'centers'
>*Repeat steps 2 and 3 until convergence:*
>
>2. Assign each point to the cluster 'center' closest in euclidian distance
>3. Cluster 'centers' are replaced by the average of all points assigned to them

Convergence here means that the change applied during each repetition of step 4 stops happening (the algorithm has 'settled' on the cluster centers).

#### K-Means Pros & Cons

Pros | Cons
--------- | -----------
simple to implement | Need to choose _k number_ manually
can campture clusters well, if data is clumped in fairly circular blobs | If the data features more irregularly shaped blobs, k-means will be inadequate
does not use probability | is dependent on initial values

### Gaussian Mixture Models
Where K-means is rigid due to simplicity, gaussian mixture models are more flexible.

The broad idea of GMMs is to use a gaussian distribution to represent each cluster. This allows us to consider the probability that a given point came from the distribution which represents a particular cluster.

A gaussian mixture is simply multiple gaussians combined into one function.

> Gaussian Mixture Model
>
> $p(x) = \sum_{i=1}^{K}{\phi_{i} N(x | \mu_{i}, \sigma_{i})}$
> 
> $\phi_{i}$ represents the weight applied to gaussian i *(these weights add up to 1)*
>
> $\mu_{i}$ and $\sigma_{i}$ are the mean and covariance matrix of gaussian i

**Expectation Maximization**

If the parameters of the GMM are the weight, the mean, and covariance of each gaussian, how could we figure out the ideal values of these parameters to represent our data? Doing this analytically is effectively impossible in most situations. Instead, we can try to maximize the probability that our observed data comes from a GMM with specific parameter values. This is done iteratively by repeatedly changing the values until the process converges. This is called expectation maximization.

The algorithm goes as follows:

> 1. Set the parameters  $\phi_{i}$, $\mu_{i}$,  and $\sigma_{i}$ to initial values based on the data
> 2. For each point, calculate the probability of belonging to each component gaussian
> > $\gamma_{i,k}$ = probability of $x_{i}$ belonging to component $k$
> 3. Given these probabilities, update the parameters in this order
> > $\phi_{k} = \sum_{i=1}^{N}{\frac{\gamma_{i,k}}{N}}$
> >
> > $\mu_{k} = \frac{\sum_{i=1}^{N}{\gamma_{i,k} x_{i}}}{\sum_{i=1}^{N}{\gamma_{i,k}}}$
> >
> > $\sigma_{k}^{2} = \frac{\sum_{i=1}^{N}{\gamma_{i,k}(x_{i} - \mu_{i})^{2}}}{\sum_{i=1}^{N}{\gamma_{i,k}}}$
> 
> Stop once the change in the above is smaller than some threshold

After this has been done, Baye's theorem can be used to calculate the probability that a given point belongs to each component gaussian. The one with highest probability is the one that the point is assigned to for clustering output.

#### Gaussian Mixture Models Pros & Cons

Pros | Cons
--------- | -----------
Can capture a wider variety of shapes of clusters than simpler models (*K-Means*) |  can still only capture clusters whose shape can fit into a gaussian distribution
Account for probability rather than using one hard metric as a cutoff | clusters that have more irregular shapes, GMMs might be inadequate to capture them


### DBSCAN(*Density Based Clustering*)
K-means and GMM algorithms assume the clusters being identified are convex blobs, since those can be represented by distance from a center point or by gaussian distributions around some center point. Density based clustering is a way to go about identifying clusters where the center point is less relevant, and where the contiguity of the group is what matters more.

&nbsp;

DBSCAN stands for Density Based Spatial Clustering of Applications with Noise. The overall idea is to use some pre decided density to evaluate whether a given point is part of a group or an outlier. The algorithm has two parameters: 
- $\epsilon$ : the distance from a point that will be considered its neighborhood
- $minPts$ : the minimum number of points needed in a neighborhood to define a cluster

The algorithm is as follows:
> 1. Decide on $\epsilon$ and $minPts$
> 2. Pick a random point P not yet visited
> 3. If there are $minPts$ in the neighborhood of P, create a cluster and add all points in that neighborhood to that cluster
> > For each point in the neighborhood of P, repeat step 3, adding them all to the same cluster
> > 
> > EXCEPT... if a point in the neighborhood does NOT have $minPts$ in its own neighborhood, add it to the cluster but DO NOT repeat step 3 for that point
> 4. If there are NOT $minPts$ in the neighborhood of P, label that point as noise (not part of a cluster)
> 5. Repeat steps 2 - 4 until all points have been added to a cluster or labeled as noise

Intuitively, the algorithm is traversing the data and labeling sections of it that satisfy a minimum density as clusters. The shape doesn't matter, low density is what is used to separate groups. As long as the established density is a good fit for the data, any shape of cluster can be captured.

#### DBSCAN Pros & Cons

Pros | Cons
--------- | -----------
Ability to capture clusters with irregular shapes | Clusters are assumed to have a uniform density throughout
Very effective for identifying outliers | Requires that you select a value for the density, not very necessarily straighforward


### 1.3.3 Dimension Reduction for Visualization

Dimension Reduction refers to any processes by which the dimensions of the data are limited to some target (such as 3 dimensions, which would allow for visualization). A simple way to reduce dimensions is to just remove some of the features. This can work in many contexts, by selecting the features which capture the most variance. However, in our case it might not make sense to cut out some genes arbitrarily. 

&nbsp;

**Principal Component Analysis** is a process that is used to *project* data onto a lower dimension. Without getting into the technical details, the way it works is that it finds the hyperplane along which the data varies most, and then projects the data onto that hyperplane. Each time this is done the dimensionality of the data is reduced by 1. This is repeated until the desired number of dimensions are reached.

&nbsp;

Imagine a 2D scatter plot. PCA would find the line along which the points are most spread out, and then project the points onto that line. Think of this like moving each point in the scatter plot onto the line, following the shortest distance to the line (perpendicular to it). This has essentially 'captured' as much variance as possible from the 2D data onto a single dimension.

&nbsp;

**THE CATCH**

It is very important to know that once PCA is performed on data, the resulting data is not interpretable. Our data, for example consists of gene expressions as the features. If you picked 3 genes and plotted that subset of the data, you could actually look at a point on the graph and interpret how its gene expressions differ from another point on the graph. 

Once this data is projected onto lower dimensions, the resulting axes no longer represent anything tangible, so that interpretability is not possible. The use of PCA here is purely to visualize the results of clustering, you cannot interpret anything about how the distinct clusters differ in terms of specific gene expressions by looking at the graph.

```python
def create_genes_subset(split_size=.2):
    """
        Creates a smaller dataframe from the large 'genes.csv' file based on a split from the metadata file.
        Returns a dataframe that has been transformed by log2 and the index needed for the remaining samples
        from the genes.csv file to remain independent.
    
    """
    
    metadata = pd.read_csv('../../metadata.csv')
    
    big_split, small_split = train_test_split(metadata, test_size=split_size, random_state=4)
    
    skiplines_small = np.sort(big_split.index) + 1 
    skiplines_big = np.sort(small_split.index) + 1
    
    genes_small = pd.read_csv('../../genes.csv', skiprows=skiplines_small)
    
    genes_nonAllZero = genes_small.loc[:,~genes_small.isin([0]).all(axis=0)]
    
    genes_log2_trans = np.log2(genes_nonAllZero.iloc[:,1:] + 1)
    genes_log2_trans['barcode'] = genes_small['barcode']
    
    genes_merged = pd.merge(left=small_split, right=genes_log2_trans, how='left', left_on='barcode', right_on='barcode')
    
    # Releasing memory by deleting dataframes
    del genes_small, genes_nonAllZero, genes_log2_trans, skiplines_small, skiplines_big
    
    return genes_merged, big_split, small_split
```

```python
genes_merged, big_split, small_split = create_genes_subset(split_size=.7)
```

## 3.1 Experiment Setup and Model Selection

Like explained earlier, _Classification_ is a type of supervised machine learning where we take our predictor features to then predict a specific label. There are a variety of classification models available for us in `scikit-learn` (eg. _Logisic regression, Decision Trees, Naive Bayes, etc_ ), each with it's own merits and best use cases, and it is on the researcher to figure out what models best suit both answering the question at hand and with which ones the data works best. 

For instance, let's say we heard of how useful [_Naive Bayes Classifiyers_](https://en.wikipedia.org/wiki/Naive_Bayes_classifier) were for working with "High Dimentional" data such as our gene count dataset. But if we make a quick reference to the description of the model, we'll find out that:

For example, a fruit may be considered to be an apple if it is red, round, and about 10 cm in diameter. A naive Bayes classifier considers each of these features to contribute independently to the probability that this fruit is an apple, regardless of any possible correlations between the color, roundness, and diameter features._

Since currently we don't have knowledge of the indepence of a gene count on another, it would not be recommended to use Naive Bayes as one of our classification models.

For our exaple purposes, we'll use _Logistic Regression, Random Forest, and Support Vector Machines (SVM)_ for our classification models. These are some of the most commonly used models and are always a good starting base for most research and machine learning projects.

## 3.2 Logistic Regression

The most basic of classification models is Logistic Regression, and in many cases it is used as a base model to measure how well we are performing. For python, we can follow the steps on using machine learning models discussed earlier to implement our model where, we call _sklearn_ and import `LogisticRegression`, we initiate the model, train the model, and make our predictions.

```python
# importing the logistic regression model
from sklearn.linear_model import LogisticRegression

# initiating the model
# keeping defaul hyperparamenters but setting class_weight as 'balanced' to balance the classes
logreg_cls = LogisticRegression()

# Training the model
logreg_cls.fit(X_train, y_train)

# make predictions
logreg_predicted = logreg_cls.predict(X_test)
```

Ok let's also import some ways to evaluate our models by checking metrics such as a `confusion_matrix`

```python
# importing confusion matrix to compare actual test labels vs the predicted by model
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay 

# creating labels from the '0' and '1' binary labels for easier display
y_test_labels = np.where(y_test, 'BRAD', 'Not BRAD')
logreg_pred_labels = np.where(logreg_predicted, 'BRAD', 'Not BRAD')

def plot_cm(y_true, y_pred, figsize=(10,10)):
    """
        Plot for the confusion matrix taking in the labels for the real values and predicted
    """
    
    # implementing the confusion matrix
    cm = confusion_matrix(y_true, y_pred, labels=np.unique(y_true))
    cm_sum = np.sum(cm, axis=1, keepdims=True)
    cm_perc = cm / cm_sum.astype(float) * 100
    annot = np.empty_like(cm).astype(str)
    nrows, ncols = cm.shape
    for i in range(nrows):
        for j in range(ncols):
            c = cm[i, j]
            p = cm_perc[i, j]
            if i == j:
                s = cm_sum[i]
                annot[i, j] = '%.1f%%\n%d/%d' % (p, c, s)
            elif c == 0:
                annot[i, j] = ''
            else:
                annot[i, j] = '%.1f%%\n%d' % (p, c)
    cm = pd.DataFrame(cm, index=np.unique(y_true), columns=np.unique(y_true))
    cm.index.name = 'Actual'
    cm.columns.name = 'Predicted'
    fig, ax = plt.subplots(figsize=figsize)
    sns.heatmap(cm, cmap= "Blues", annot=annot, fmt='', ax=ax)
    
plot_cm(y_test_labels, logreg_pred_labels)
plt.savefig('../3-Outputs/logreg_untuned_conmatrix.png')

# releasing memory allocation
del logreg_pred_labels
```

Ok wow! We perfectly predicted which samples were _BRAD_ and which were not! Amazing indeed...
Now this type for perfect predictic capabilities should raise concerns about our model. Recall that when something like this occurs, this is known as __Overfitting__ . Moreover, we have a _warning message_ from the sklearn model _"ConvergenceWarning: lbfgs failed to converge (status=1):
STOP: TOTAL NO. of ITERATIONS REACHED LIMIT"_ . Our Logistic Regression model did not converged and as a result the paramenters simply reside somewhere on the boundaries.

The message also gives options as to how to _possibly_ achive convergence, either increase the number of iterations the model performs or increase the number of samples in our data. We're going to take this opportunity to explore a way to address such a high feature/dimentional dataset. 

### Principal Component Analysis _(PCA)_

As mentioned earlier, _Principar Component Analysis_ is fundamentally a dimention reduction algorithm, that can be useful for noice filtering, selecting important features, and visualizations of high dimensional data.

PCA's algorithm computes the _principal axes_ in our dataset, which basically means it looks for the maximum variance in our features or variables. The algorithm runs through all the combination of features and calculates vectors for the max variance, basically projecting our original dataset into a plane defined by these vectors. 

We can call the _PCA_ algorithm through sklearn, select the number of components we would like our dataset to be reduced to (by default 2), transform our data, and rerun our Logistic Regression model with this new subset.

```python
# importing PCA from sklearn
from sklearn.decomposition import PCA

# implementing PCA
pca = PCA(n_components=2) #assume to keep 2 components, but doesn't matter
pca_X_train = pd.DataFrame(pca.fit_transform(X_train))
pca_X_test = pd.DataFrame(pca.transform(X_test))

# model instanciation, changing the class_weight hyperparameter to account for unbalanced classes
lrpca_cls = LogisticRegression(class_weight='balanced') 
# training model
lrpca_cls.fit(pca_X_train, y_train)

# making predictions 
lr_pca_pred = lrpca_cls.predict(pca_X_test)

lrpca_pred_labels = np.where(lr_pca_pred, 'BRAD', 'Not BRAD')

# viz for confusion matrix
plot_cm(y_test_labels, lrpca_pred_labels)
plt.savefig('../3-Outputs/logreg_tuned_conmatrix.png')
# releasing memory
del lrpca_pred_labels
```

```python
# calling in the matrics function to check how well our predictions did
from sklearn import metrics

print(metrics.classification_report(lr_pca_pred, y_test))
print("Accuracy: %3f" % (np.mean(lr_pca_pred == y_test)))
```

Ok so now that we have implemented PCA and done some hyperparamenter changes to account for the unbalance in our dataset, we have a more realistic model and predictions. 

Ok so we see that with a few tweeks we have found a better model according to our metrics.

We could even ask ourselves if us going for a PCA dimension reduction to only 2 principal conponents was appropriate. To do this we look towards how much variance is explained by the _first N PCs_.

```python
# doing PCA to check how many PCs are good enough to explain variance
pca = PCA().fit(X_train)

# setting figure parameters for size
fig_dims = (15, 7)
fig, ax = plt.subplots(1,1,figsize=fig_dims)

plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.title('PC explained variance')
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.savefig('../3-Outputs/PCA_explained_variance.png')
```

Ok this is much more interesting, we see that we need about 300 PCs to account for about 80% of the variance in our dataset, while keeping it in the 10s would acounts to about 30%-40%. A lower dimensionality like 2 and 3 is easy to visualize so we might just want to look at something around those values if for visualization purposes, but lets also calculate how much would the model improve at predicting labels if we increase the PC to 3.

First, let's use the PCA we have with 2 Principal Components to set up a visual and see how well the data has been split between samples that are 'BRAD' and those that are 'NOT BRAD'

```python
# setting figure parameters for size
fig_dims = (15, 6)
fig, ax = plt.subplots(1,2,figsize=fig_dims)
# Data for a axis from PCs
xline = pca_X_test.iloc[:, 0]
yline = pca_X_test.iloc[:, 1]
# plotting
sns.scatterplot(x=xline, y=yline,hue=lr_pca_pred, cmap='colorblind', ax=ax[0], palette='colorblind', legend='full')
sns.scatterplot(x=xline, y=yline,hue=np.array(y_test), cmap='colorblind', ax=ax[1], palette='colorblind', legend='full')

ax[0].set_title("Predicted")
ax[1].set_title("Actual")
ax[0].set(xlabel="PC1", ylabel="PC2")
ax[1].set(xlabel="PC1", ylabel="PC2")

fig.suptitle('Logistic Regression Prediction vs Actual - PCA visualized', fontsize=16)
plt.legend(loc='upper right')
plt.show()
plt.savefig('../3-Outputs/logreg_separation_PCA.png')
```

The visual shows the great power PCA can have in our data analysis. Remember that our dataset was originally of over 50K genes or features and we then projected the data and reduced the dimension of our dataset to only 2 dimensions... and it did a fair job at showing a separation for our 2 classes, but it does have some of the inaccuracies when compared to the actual values, meaning our data has more complexity than just the 2 dimensions we used.

Ok now let's proceed to redo our PCA procedure, but for 3 Principal Components now and see how much our Classifier improves.

```python
# Rerunning the PCA procedure
pca = PCA(n_components=3) #change to 3 components
newdf_train = pca.fit_transform(X_train)
newdf_valid = pca.transform(X_test)

features_train = pd.DataFrame(newdf_train)
features_valid = pd.DataFrame(newdf_valid)  
del newdf_train, newdf_valid

lrpca_cls = LogisticRegression(class_weight='balanced') 
lrpca_cls.fit(features_train, y_train)
lr_pca_pred = lrpca_cls.predict(features_valid)

lrpca_pred_labels = np.where(lr_pca_pred, 'BRAD', 'Not BRAD')

plot_cm(y_test_labels, lrpca_pred_labels)
plt.savefig('../3-Outputs/logreg_tuned_conmatrix.png')
print(metrics.classification_report(lr_pca_pred, y_test))
print("Accuracy: %3f" % (np.mean(lr_pca_pred == y_test)))
```

AWESOME!!! Increasing the PCs to 3 increased our performance metrics grately, going from about mid 70%s to close to mid 80%s in most metrics. This might not always be the case when just increasing a value by 1 PC, but this is the part of the Machine Learning process where the _art of science_ cames in and we _figure out how to tune_ our hyperparaments for better performance.

### RandomForests Models

_"You can't see the forest for the trees"_ , a quote suggesting we might miss the big picture if we focus too much in the small detail. Not to suggest that small details aren't important, but a nice way to say for us _"Hey, why just use one tree model, when we can look at a whole bunch of them"_ .

___Random Forest___ models are an emsemble of Tree models, where we define how many trees we want. These can be used not only for classification but also in regression purposes by running a multitude of these tree models and outputting the class that is the average prediction of the individual trees. One added bonus for our _cancer type classification_ purposes that Randon Forest models bring, is that they allow us to do multiple classes and not just stick to the binary classification for the previous two models. Like we discussed earlier, one of the "problems" or disadvantages with _Tree_ models, is the fact that it can feel like a blackbox where you _put stuff in and stuff that works pops out_ . There is no easy way around this fact, but something that might help us understand to some extend what the model is doing, we can try to visualize a the levels or sections how the model places it's predictions. 

We'll first use PCA again to reduce the dimension of our data for visualization purposes, and then we'll create this function to help visualize the classification of the model.

```python
# implementing PCA
pca = PCA(n_components=3) 
pca_X_train = pd.DataFrame(pca.fit_transform(X_train))
pca_X_test = pd.DataFrame(pca.transform(X_test))
```

```python
def visualize_classifier(model, X, y,X_test, y_test, ax=None):
    """
        Creates a visualization grid space for the areas a data point fall into
    """
    ax = ax or plt.gca()
    # Plot the training points
    scatter = ax.scatter(X_test[:, 0], X_test[:, 1], c=y_test, s=30, cmap='rainbow', edgecolors= "black",
               clim=(y.min(), y.max()), zorder=3)
    ax.axis('tight')
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    # fit the estimator
    model.fit(X, y)
    xx, yy = np.meshgrid(np.linspace(*xlim, num=200),
                         np.linspace(*ylim, num=200))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

    # Create a color plot with the results
    n_classes = len(np.unique(y))
    contours = ax.contourf(xx, yy, Z, alpha=0.3,
                           levels=np.arange(n_classes + 1) - 0.5,
                           cmap='rainbow')
    
    # produce a legend with the unique colors from the scatter
    legend1 = ax.legend(*scatter.legend_elements())
    ax.add_artist(legend1)

    ax.set(xlim=xlim, ylim=ylim)
```

```python
# setting figure parameters for size
fig_dims = (15, 6)
fig, ax = plt.subplots(2,3,figsize=fig_dims)
# Data for a axis from PCs
xline = pca_X_test.iloc[:, 0]
yline = pca_X_test.iloc[:, 1]
zline = pca_X_test.iloc[:, 2]
# plotting predicted
sns.scatterplot(x=xline, y=yline,hue=rf_pred, cmap='colorblind', ax=ax[0,0], palette='colorblind', legend='full')
sns.scatterplot(x=xline, y=zline,hue=rf_pred, cmap='colorblind', ax=ax[0,1], palette='colorblind')
sns.scatterplot(x=yline, y=zline,hue=rf_pred, cmap='colorblind', ax=ax[0,2], palette='colorblind')

# plotting actual
sns.scatterplot(x=xline, y=yline,hue=np.array(y_test), cmap='colorblind', ax=ax[1,0], palette='colorblind', legend='full')
sns.scatterplot(x=xline, y=zline,hue=np.array(y_test), cmap='colorblind', ax=ax[1,1], palette='colorblind')
sns.scatterplot(x=yline, y=zline,hue=np.array(y_test), cmap='colorblind', ax=ax[1,2], palette='colorblind')


ax[0,1].set_title("Predicted")
ax[1,1].set_title("Actual")
ax[0,0].set(xlabel="PC1", ylabel="PC2")
ax[0,1].set(xlabel="PC1", ylabel="PC3")
ax[0,2].set(xlabel="PC2", ylabel="PC3")
ax[1,0].set(xlabel="PC1", ylabel="PC2")
ax[1,1].set(xlabel="PC1", ylabel="PC3")
ax[1,2].set(xlabel="PC2", ylabel="PC3")
plt.tight_layout(pad=1)
plt.subplots_adjust(top=0.9)


fig.suptitle('Random Forest Prediction vs Actual - PCA visualized', fontsize=16)
plt.savefig('../3-Outputs/rForest_separation_PCA.png')
plt.show()
```

```python
# setting figure parameters for size
fig_dims = (17, 4)
fig, ax = plt.subplots(1,3,figsize=fig_dims)
visualize_classifier(rf_model, np.array(pca_X_train)[:,:2], y_train,np.array(pca_X_test)[:,:2], y_test,ax=ax[0])
visualize_classifier(rf_model, np.array(pca_X_train)[:,[0,2]], y_train,np.array(pca_X_test)[:,[0,2]], y_test,ax=ax[1])
visualize_classifier(rf_model, np.array(pca_X_train)[:,1:], y_train,np.array(pca_X_test)[:,1:], y_test,ax=ax[2])

ax[0].set(xlabel="PC1", ylabel="PC2")
ax[1].set(xlabel="PC1", ylabel="PC3")
ax[2].set(xlabel="PC2", ylabel="PC3")

fig.suptitle('Random Forest Prediction Space', fontsize=16)
plt.savefig('../3-Outputs/rForest_Decision_space.png')
plt.show()
```

We see how the predicted vs the actual values for the random forest did extremely well, as expected from the confusion matrix. However, it is also very interesting how the "predicted decision space" (as we are calling it for this visual), shows that not all points in a specific _color_ for a set of _PCs_, end up with that label. This further implies that there is more complexity than can optimally be described with our PCA values, which is to be expected. Nonetheless, this provides a fairly good visual aid for us to understand the "blackbox" magic of a random forest model. 

## 3.4 Support Vector Machines _(SVM)_

The last supervised model we'll use to accomplish our goal of _Classifying Cancer types by using gene count data_ is a __Support Vector Machines (SVM)__ model. 


```python
from sklearn.svm import SVC 
from sklearn.pipeline import make_pipeline

pca = PCA(n_components=100, svd_solver='randomized', whiten=True, random_state=42)
svc = SVC(kernel='rbf', class_weight='balanced')
model = make_pipeline(pca, svc)
```

SVMs provide a wide range of tunning so we can optimize our model and at this point we would start to wonder about how to tune our [___hyperparameters___](https://en.wikipedia.org/wiki/Hyperparameter_optimization) (think like a 'Knob' on a radio), which remember are _'paramenters'_ intrisic to our model that we can change around, to improve our model. This could be by either changing the weight we give our classes or a number of other paramenters. For our example, let's focus on changing the '`svs_C` and `svc_gamma` parameters and do a _Grid Search_ for a few values. 

[___Grid Search___](https://en.wikipedia.org/wiki/Hyperparameter_optimization#Grid_search) is a method where we provide a set of parameters for our model to compute over and where we would optain the _optimized_ value for the set of parameters (or _Grid_ ) that we provided. For our grid search, since we have an idea already of the ratio of the binary categories (yes BRAD and Not BRAD), we'll try the values `{'svc__C': [1, 5, 10, 50]}` and `{'svc__gamma': [0.0001, 0.0005, 0.001, 0.005]}` that are close to the value and see which might _optimize_ our svm model. 

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'svc__C': [1, 5, 10, 50],
              'svc__gamma': [0.0001, 0.0005, 0.001, 0.005]}
grid = GridSearchCV(model, param_grid)

%time grid.fit(X_train, y_train)
print(grid.best_params_)
```

Ok so we got that the _Optimized_ model value came from the `{'svc__C': 1, 'svc__gamma': 0.0005}`, great we can plug it in for our weights, get the best model and keep moving on... however, if for some reason we had noticed that the hyperparameter are one of the _boundaries_ in our grid, that is the "edge" values in our lists, this would mean we need to expand our grid for a better value. For our example we can proceed with the best values optained by the _Grid Search_ .   
_NOTE: As we can see Grid Search can take quite some time to compute. It is best if we have some idea of where to confine our search parameters for optimal time use._

```python
# setting best model from grid search parameters
svc_model = grid.best_estimator_
# predicting on test values
svc_pred = svc_model.predict(X_test)
```

```python
cond_test = [ y_test == 1, y_test == 2, y_test == 3, y_test == 4, y_test == 5 ]
cond_pred = [ svc_pred == 1, svc_pred == 2, svc_pred == 3, svc_pred == 4, svc_pred == 5 ]
choices = [ 'PRAD', 'BRAD', 'COAD', 'LUAD', 'KIRC']
svc_test_labels = np.select(cond_test, choices, default=np.nan)
svc_pred_labels = np.select(cond_pred, choices, default=np.nan)

print(metrics.classification_report(svc_pred, y_test))
plot_cm(svc_test_labels, svc_pred_labels)
plt.title('SVMt Confusion Matrix for Cancer Types')
plt.savefig('../3-Outputs/SVM_conmatrix.png')
```

# 4. Unsupervised Machine Learning

In the following section we will make use of Scikit-Learn's clustering functions to perform clustering on our data with the three algorithms covered in the background section.

Similarly as to how we use sklearn functions for the supervised models, there is a structured process that can be followed when using the unsupervised models for sklearn.

> The steps to follow again would be:
- Instantiate the model
    - ex: `kmeans = KMmeans(*parameters*)`
- 'Fit' the model by training on data
    - `kmeans.fit(*data*)`  
    _Note:  Since we are clustering without the explicit intent to make predictions with our models, we don't have to split our data into 'training' and 'testing' sets. **We can simply train the models on our entire data**._
- Call functions to see the 'observations' or results from the model
    - ex: `kmeans.labels_`  
    _Note: Almost every scikitlearn model has a 'labels_' attribute. We can use this after training to access the labels that have been generated so we can use them to color our visualizations_

## 4.1 K-Means

First let's try K-Means clustering. The function for K-Means from Scikit-Learn has many parameters, the most important one being the nubmer of clusters. This is how many centerpoints the algorithm will start (and end) with. 

Your selection of number of clusters K should reflect how many groups you think the data represents.

Let's use the gene expression data as the basis for clustering. Afterwards we will see if the results of clustering line up with any categorical data, such as cancer label.

To decide how many clusters to use, let's base it on the number of cancer types. This assumes that clusters in the data would correspond to cancer types. There are 5 different types in this data.

```python
# importing desired model
from sklearn.cluster import KMeans

# Call sklearn k-means clustering for 5 clusters
kmeans = KMeans(n_clusters=5).fit(X)

# Use PCA to reduce features for visualization
pca_genes = PCA(n_components=3).fit_transform(X)
```

```python
# Use Matplotlib to visualize in 3D
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
fig.set_figheight(7)
fig.set_figwidth(18)

# Kmeans clusters by using kmeans.labels as color
ax = fig.add_subplot(1,2,1, projection='3d')
ax.set_title("K-Means Clustering on Gene Expression")
ax.scatter(pca_genes[:,0], pca_genes[:,1], pca_genes[:,2], c=kmeans.labels_, cmap='tab10')

# Do the same but color by cancer type
ax = fig.add_subplot(1,2,2, projection='3d')
labels = {"PRAD": 0, "BRCA": 1, "COAD": 2, "LUAD": 3, "KIRC" : 4}
ax.set_title("Gene Expressions Projected to 3D Labeled by Cancer Type")
ct_scatter = ax.scatter(pca_genes[:,0], pca_genes[:,1], pca_genes[:,2], 
                        c=genes_merged['cancer_type'].map(labels), cmap='tab10')

# produce a legend with the unique colors from the scatter
ax.add_artist(ax.legend(*ct_scatter.legend_elements()))

plt.tight_layout(pad=1)
plt.subplots_adjust(top=0.9)

fig.suptitle('KMeans Cluster vs Known Cancer Group Clusters', fontsize=16)
plt.savefig('../3-Outputs/Kmeans_Cluster.png')
plt.show()
```

#### K-Means Results
As visible from the above visualizations, the K-means clustering does capture groups that appear to be clusters (they are actually separate blobs). Additionally, 3 of the clusters identified line up very closely with cancer types, which indicates that the clusters can actually capture something useful in the data.


## 4.2 Gaussian Mixture Models
Now let's do the exact same thing but with a GMM. In this case, the GMM does not assign a cluster label to points directly, but once the model is generated, the highest probable gaussian for each point can be assigned as its cluster.

We use the dimension-reduced data as input for the clustering because the full data demands gigabytes of memory for the algorithm to run. As you read in the background section, GMM involves significantly more complex math than K-means.

```python
# import GMM 
from sklearn.mixture import GaussianMixture
# Set up the GMM using scikitlearn
gmm = GaussianMixture(n_components=5)
# fit
gmm.fit(pca_genes)
```

```python
# Use the GMM to assign a label to each point
labels = gmm.predict(pca_genes)

# Visualize the results
fig = plt.figure()
fig.set_figheight(7)
fig.set_figwidth(18)

ax = fig.add_subplot(121, projection='3d')
ax.set_title("GMM Clustering on Gene Expression")
ax.scatter(pca_genes[:,0], pca_genes[:,1], pca_genes[:,2], c=labels, cmap='tab10')

# Do the same but color by cancer type
ax = fig.add_subplot(1,2,2, projection='3d')
labels = {"PRAD": 0, "BRCA": 1, "COAD": 2, "LUAD": 3, "KIRC" : 4}
ax.set_title("Gene Expressions Projected to 3D Labeled by Cancer Type")
ct_scatter = ax.scatter(pca_genes[:,0], pca_genes[:,1], pca_genes[:,2], 
                        c=genes_merged['cancer_type'].map(labels), cmap='tab10')

# produce a legend with the unique colors from the scatter
ax.add_artist(ax.legend(*ct_scatter.legend_elements()))

plt.tight_layout(pad=1)
plt.subplots_adjust(top=0.9)

fig.suptitle('GMM Cluster vs Known Cancer Group Clusters', fontsize=16)
plt.savefig('../3-Outputs/GMM_Cluster.png')
plt.show()
```

#### Gaussian Mixture Model Result
The result above displays one of the strengths of GMMs over K-Means. The top right (purple) cluster is able to capture the more elongated shape of the underlying group. Overall the results are good and indicate that GMMs are capable of capturing the cancer type groups in this data.


## 4.3 DBSCAN
Now let's try density based clustering on our data. Immediately we are met with one of the challenges of applying DBSCAN: we must decide on what distance to consider around each point, and how many points to require within that distance. These starting parameters are the most important part of applying DBSCAN correctly, as they must be specifically chosen for your data.

To see an example of how much this matters, let's first run DBSCAN with some arbitrary values.

```python
# import model
from sklearn.cluster import DBSCAN
# Set up DBSCAN and run it on our data
dbscan = DBSCAN(eps=1, min_samples=10)
# fit model
dbscan.fit(pca_genes)
```

```python
# Visualize the results
labels = dbscan.labels_

fig = plt.figure()
fig.set_figheight(10)
fig.set_figwidth(10)
ax = fig.add_subplot(111, projection='3d')
ax.set_title("DBSCAN Clustering on Gene Expression")
ax.scatter(pca_genes[:,0], pca_genes[:,1], pca_genes[:,2], c=labels)
```

#### What went wrong?
Clearly DBSCAN did not work, and at first glance you might wonder if it even did anything. How can all the points be in one cluster. If we take a look at the labels assigned to each point by DBSCAN, they are all -1. This is the label for noise. All of our points are being considered as noise by the model. This means a distance of 1 around each point is too small to capture the clusters here.

```python
# Set up DBSCAN with new parameters
dbscan = DBSCAN(eps=3, min_samples=10)
# fit model
dbscan.fit(pca_genes)
```

```python
# Visualize the results
labels = dbscan.labels_

# Visualize the results
fig = plt.figure()
fig.set_figheight(7)
fig.set_figwidth(18)

ax = fig.add_subplot(1,2,1, projection='3d')
ax.set_title("DBSCAN Clustering on Gene Expression")
ax.scatter(pca_genes[:,0], pca_genes[:,1], pca_genes[:,2], c=labels, cmap='tab10')

# Do the same but color by cancer type
ax = fig.add_subplot(1,2,2, projection='3d')
label = {"PRAD": 0, "BRCA": 1, "COAD": 2, "LUAD": 3, "KIRC" : 4}
ax.set_title("Gene Expressions Projected to 3D Labeled by Cancer Type")
ct_scatter = ax.scatter(pca_genes[:,0], pca_genes[:,1], pca_genes[:,2], 
                        c=genes_merged['cancer_type'].map(label), cmap='tab10')

# produce a legend with the unique colors from the scatter
ax.add_artist(ax.legend(*ct_scatter.legend_elements()))

plt.tight_layout(pad=1)
plt.subplots_adjust(top=0.9)

fig.suptitle('DBSCAN Cluster vs Known Cancer Group Clusters', fontsize=16)
plt.savefig('../3-Outputs/DBSCAN_Cluster.png')
plt.show()
```

#### DBSCAN Results
With new initial parameters, DBSCAN can capture some clusters in our data, but the result is not as close to the underlying groups as the previous methods. The purple points are labeled as noise (not part of a cluster). This demonstrates one of the weaknesses of DBSCAN, and why we should evaluate our data before picking a clustering algorithm. As noted in the background, DBSCAN will suffer when clusters have significant changes in density within them. The top right cluster captured by GMM was able to include some of the far off points that have the same cancer type (the cluster lines up more closely with what is going on in our data). DBSCAN labels many of these as noise. 

We could have known that DBSCAN would be a bad model by looking at the visualization of the data colored by cancer type. The underlying groups were both stretched out and not even close to uniform in density.
